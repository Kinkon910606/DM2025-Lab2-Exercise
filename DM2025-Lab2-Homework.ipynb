## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
import pandas as pd
import os
import json
os.getcwd()

### Load Data
print("\n === Load Data... ===")
dataidentification = pd.read_csv('./dm-lab-2-private-competition/data_identification.csv')
emotion = pd.read_csv('./dm-lab-2-private-competition/emotion.csv')
samplesubmission = pd.read_csv('./dm-lab-2-private-competition/samplesubmission.csv')

with open('./dm-lab-2-private-competition/final_posts.json', 'r', encoding='utf-8') as f:
    final_posts = json.load(f)
print(f"dataidentification shape: {dataidentification.shape}, emotion shape: {emotion.shape}, samplesubmission shape: {samplesubmission.shape}")

### Split Data into Train and Test Sets
print("\n === Split Data into Train and Test Sets... ===")
train = dataidentification[dataidentification['split'] == 'train']
test = dataidentification[dataidentification['split'] == 'test']
print(f"Train shape: {train.shape}, Test shape: {test.shape}")

### Add emotion labels to train set and sample submission structure to test set
print("\n === Add emotion labels to train set and sample submission structure to test set... ===")
train = pd.merge(emotion, train, on='id', how='left')
test = pd.merge(samplesubmission, test, on='id', how='left')

### Add text data from final_posts JSON to train and test sets
print("\n === Add text data from final_posts JSON to train and test sets... ===")
final_posts = {
    item.get('root', {}).get('_source', {}).get('post', {}).get('post_id'):
    item.get('root', {}).get('_source', {}).get('post', {}).get('text', "")
    for item in final_posts
    if item.get('root', {}).get('_source', {}).get('post', {}).get('post_id') is not None
}
train['text'] = train['id'].apply(lambda x: final_posts.get(str(x), ""))
test['text'] = test['id'].apply(lambda x: final_posts.get(str(x), ""))
```

## 2. Feature Engineering Steps

#### Text Preprocessing

```python
import nltk
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_tokenizer(text):
    return [lemmatizer.lemmatize(tok) for tok in nltk.word_tokenize(text.lower())]
```

```python

import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

import emoji

print('\n === Text Preprocessing... ===')
# 第一次使用需要下載 stopwords
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')


stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower() # 轉換為小寫
    text = emoji.replace_emoji(text, replace='')           
    text = re.sub(r"http\S+|www\S+", " ", text)           
    text = re.sub(r"@\w+", " ", text)                    
    text = re.sub(r"#(\w+)", r"\1", text)                 
    text = re.sub(r"[^a-z0-9'!?.,\s]", " ", text)         
    text = re.sub(r"\s+", " ", text).strip()
   
    words = text.split()  # 分詞並移除停用詞
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

# 將前處理應用到訓練集和測試集的 text 欄位
train['processed_text'] = train['text'].apply(preprocess_text)
test['processed_text'] = test['text'].apply(preprocess_text)
```

```python
train.to_csv("./dm-lab-2-private-competition/train_cleaned.csv", index=False)
test.to_csv("./dm-lab-2-private-competition/test_cleaned.csv", index=False)
```

#### train_test_split

```python
from sklearn.model_selection import train_test_split

X_train = train['processed_text']
y_train = train['emotion']
X_test = test['processed_text']
y_test = test['emotion']


X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

print(f"X_train shape: {X_train.shape} >>> type {type(X_train)} ")
print(f"X_val shape: {X_val.shape} >>> type {type(X_val)} ")
print(f"X_test shape: {X_test.shape} >>> type {type(X_test)} ")
print(f"y_train shape: {y_train.shape} >>> type {type(y_train)} ")
print(f"y_val shape: {y_val.shape} >>> type {type(y_val)} ")
print(f"y_test shape: {y_test.shape} >>> type {type(y_test)} ")
```

```python
y_train_label = y_train.map({
    'anger': 0,
    'disgust': 1,
    'fear': 2,
    'joy': 3,
    'sadness': 4,
    'surprise': 5
})

y_val_label = y_val.map({
    'anger': 0,
    'disgust': 1,
    'fear': 2,
    'joy': 3,
    'sadness': 4,
    'surprise': 5
})
```

```python
### Count Vectorizer 
from sklearn.feature_extraction.text import CountVectorizer

BOW_vectorizer = CountVectorizer(max_features=1000)

# fitting !
BOW_vectorizer.fit(X_train) 

# transform
train_BOW_features = BOW_vectorizer.transform(X_train)
val_BOW_features = BOW_vectorizer.transform(X_val)
test_BOW_features = BOW_vectorizer.transform(X_test)
```

#### TF-IDF Vectorization

```python

### TF-IDF Vectorization
# print('\n === TF-IDF Vectorization... ===')
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk import word_tokenize

nltk.download('punkt')  # 第一次使用需要

vectorizer = TfidfVectorizer(
    max_features=100000,
    ngram_range=(1, 2),
    # min_df=2,
    # max_df=0.7,
    sublinear_tf=True,
    # tokenizer=word_tokenize,
    tokenizer=lemmatize_tokenizer,
    token_pattern=None  # 提供自訂 tokenizer 時要關閉內建 token pattern
)

# fitting !
vectorizer.fit(train['processed_text'])

# transform
train_TFIDF_features = vectorizer.transform(X_train)
val_TFIDF_features = vectorizer.transform(X_val)
test_TFIDF_features = vectorizer.transform(X_test)
```

## 3. Model Implementation Steps

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

def plot_cm(y_train, y_pred_train, y_test, y_pred_test):
    labels = np.unique(np.concatenate([y_train.values, y_pred_train, y_test.values, y_pred_test]))

    cm_train = confusion_matrix(y_train, y_pred_train, labels=labels)
    cm_test  = confusion_matrix(y_test,  y_pred_test,  labels=labels)

    fig, axes = plt.subplots(1, 2, figsize=(18, 7))

    sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', ax=axes[0],
                xticklabels=labels, yticklabels=labels)
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('True')
    axes[0].set_title('Confusion Matrix - Train Set')
    axes[0].tick_params(axis='x', rotation=45)

    sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', ax=axes[1],
                xticklabels=labels, yticklabels=labels)
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('True')
    axes[1].set_title('Confusion Matrix - Test Set')
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()
    
def result(true, pred):
    acc = accuracy_score(y_true=true, y_pred=pred)
    score = f1_score(y_true=true, y_pred=pred, average='weighted')
    print(f"acc={round(acc,2)} >>> f1-score={round(score,2)}")

```
