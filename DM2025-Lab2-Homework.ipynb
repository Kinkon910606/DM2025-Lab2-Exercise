# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
import pandas as pd
import os
import json
os.getcwd()

### Load Data
print("\n === Load Data... ===")
dataidentification = pd.read_csv('./dm-lab-2-private-competition/data_identification.csv')
emotion = pd.read_csv('./dm-lab-2-private-competition/emotion.csv')
samplesubmission = pd.read_csv('./dm-lab-2-private-competition/samplesubmission.csv')

with open('./dm-lab-2-private-competition/final_posts.json', 'r', encoding='utf-8') as f:
    final_posts = json.load(f)
print(f"dataidentification shape: {dataidentification.shape}, emotion shape: {emotion.shape}, samplesubmission shape: {samplesubmission.shape}")

### Split Data into Train and Test Sets
print("\n === Split Data into Train and Test Sets... ===")
train = dataidentification[dataidentification['split'] == 'train']
test = dataidentification[dataidentification['split'] == 'test']
print(f"Train shape: {train.shape}, Test shape: {test.shape}")

### Add emotion labels to train set and sample submission structure to test set
print("\n === Add emotion labels to train set and sample submission structure to test set... ===")
train = pd.merge(train, emotion, on='id', how='left')
test = pd.merge(test, samplesubmission, on='id', how='left')

### Add text data from final_posts JSON to train and test sets
print("\n === Add text data from final_posts JSON to train and test sets... ===")
final_posts = {
    item.get('root', {}).get('_source', {}).get('post', {}).get('post_id'):
    item.get('root', {}).get('_source', {}).get('post', {}).get('text', "")
    for item in final_posts
    if item.get('root', {}).get('_source', {}).get('post', {}).get('post_id') is not None
}
train['text'] = train['id'].apply(lambda x: final_posts.get(str(x), ""))
test['text'] = test['id'].apply(lambda x: final_posts.get(str(x), ""))
```

```python
train.groupby('emotion').size()
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section

### Text Preprocessing
import re
from nltk.corpus import stopwords
import nltk

print('\n === Text Preprocessing... ===')
#  stopwords
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower() # 
    text = re.sub(r'[^a-zA-Z\s]', '', text) # 
   
    words = text.split()  # 
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

#  text 
train['processed_text'] = train['text'].apply(preprocess_text)
test['processed_text'] = test['text'].apply(preprocess_text)

### TF-IDF Vectorization
print('\n === TF-IDF Vectorization... ===')
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk
from nltk import word_tokenize

nltk.download('punkt')  # 


vectorizer = TfidfVectorizer(
    max_features=20000,
    ngram_range=(1, 4),
    min_df=2,
    max_df=0.7,
    sublinear_tf=True,
    tokenizer=word_tokenize,
    token_pattern=None  #  tokenizer  token pattern
)

vectorizer.fit(train['processed_text'])
X_train = vectorizer.transform(train['processed_text'])
X_test = vectorizer.transform(test['processed_text'])

# 
y_train = train['emotion']
y_test = test['emotion']

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42, k_neighbors=10, sampling_strategy='auto')
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
```

## 3. Model Implementation Steps

```python

#// Method : Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,       # 
    max_depth=15,           # 
    min_samples_split=5,     # 
    min_samples_leaf=4,        # 
    random_state=42,
    max_features='sqrt',
    class_weight='balanced' 
    )

## training!
model = model.fit(X_train_balanced, y_train_balanced)

## predict!
y_train_pred = model.predict(X_train_balanced)
y_test_pred = model.predict(X_test)
```

```python
### calculate accuracy
from sklearn.metrics import accuracy_score

acc_train = accuracy_score(y_true=y_train_balanced, y_pred=y_train_pred)
acc_test = accuracy_score(y_true=y_test, y_pred=y_test_pred)

print('training accuracy: {}'.format(round(acc_train, 2)))
print('testing accuracy: {}'.format(round(acc_test, 2)))
```

```python
### calculate f1-score
from sklearn.metrics import f1_score
f1_train = f1_score(y_true=y_train_balanced, y_pred=y_train_pred, average='weighted')
f1_test = f1_score(y_true=y_test, y_pred=y_test_pred, average='weighted')
print('training f1-score: {}'.format(round(f1_train, 2)))
print('testing f1-score: {}'.format(round(f1_test, 2)))
```

```python
### confusion matrix - train set
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(y_train_balanced, y_train_pred)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
### confusion matrix - test set
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Test Set')
plt.show()
```
